## 求出数组中除了最大值和最小值后的平均数

```python
import sys
def findAve(nums):
    min = sys.maxsize
    max = -sys.maxsize
    sum=0
    for i in nums:
        if i<min:
            min=i
        if i>max:
            max=i
        sum+=i
    return (sum-min-max)/(len(nums)-2)
```

```python
import random
import numpy as np
nums = np.random.rand(100).tolist()
ans = findAve(nums)
ans
```
